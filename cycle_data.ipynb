```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bike_sharing = pd.read_csv("day.csv")
```

```python
bike_sharing.dtypes
```

Conversion of dteday column from obj to datetime64[ns]

```python
bike_sharing["dteday"] = pd.to_datetime(bike_sharing["dteday"])
```

```python
bike_sharing.dtypes
```

```python
bike_sharing
```

plotting the relation between dateday and total number of bikes rented

```python
plt.figure(figsize=(8, 6))  

plt.plot(bike_sharing["dteday"], bike_sharing["cnt"])
plt.xticks(rotation = 90)
plt.xlabel ("Date")
plt.ylabel ("Bikes Rented")

plt.show
```

plotting the relation between dateday and temperature

```python
plt.figure(figsize=(8, 6))  

plt.plot(bike_sharing["dteday"], bike_sharing["temp"])
plt.xticks(rotation = 90)

plt.show
```

finding the relation between temperature and total number of bikes rented

```python
plt.figure(figsize=(8, 6))  

plt.scatter(bike_sharing["temp"], bike_sharing["cnt"])
plt.xlabel ("Temperature")
plt.ylabel ("Bikes Rented")
```

finding corelation between temperature and total number of bikes rented 
which is 62.7%
that means the temperature is the main factor in the increase or decrease of number of bikes rented

```python
bike_sharing["temp"].corr(bike_sharing["cnt"])
```
